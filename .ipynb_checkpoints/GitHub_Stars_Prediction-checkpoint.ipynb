### Task 1.2 Load the Data

To avoid any problems with GitHub blocking us from downloading the data many times, we have downloaded and saved all the html files in the [data](data) folder. We will be treating the data folder as our website to be scraped. The path to data folder is stored in `DATA_PATH` variable.

In the data folder the first 10 pages of highly starred repositories are saved as `searchPage1.html`,`searchPage2.html`,`searchPage3.html` ... `searchPage10.html`

Now we will read these html files in python and create a soup object. This is a two step process:
 * Read the text in the html files
 * Create the soup from the files that you've read.

````python
#Reading Search pages and Storing in Soup
list = []
for x in os.listdir(DATA_PATH):
    if x.endswith(".html"):
        list.append(x)
soup = BeautifulSoup()
for x in list:
    soup.append(BeautifulSoup(open(DATA_PATH + "/" + x, encoding='utf8'), "html.parser"))
````

### 1.3. Extract Data

Extracting the following data for each repository, and creating a Pandas Dataframe with a row for each repository and a column for each of these datums. 

1. The name of the repository
+ The primary language (there are multiple or none, if multiple, use the first one, if none, use "none")
+ The number of watches
+ The number of stars
+ Number of forks
+ Number of commits
+ Number of branches
+ Number of contributors
+ Number of issues
+ Length of readme file.

Here's an example for one repository on how that data will look like: 
```python
{'name': 'Python-100-Days',
'language': 'Jupyter Notebook',
'stars': '78068',
'watches': '4822',
'forks': '30979',
'commits': 296,
'branches': 1,
'contributors': 12,
'issues': '224',
'readme': 11159}
```

Note that we will first extract links from the soup that we scraped earlier. When we extract the link for the repository it will be a path to the stored html page for the repository. We will use this path to read the file and extract the above information.

````python
#Scrapping Every Repo Page
repo = []
repo_soup = BeautifulSoup()
i = 0
names = []
languages = []
stars = []
watches = []
forks = []
commits = []
branches = []
contributors = []
issues = []
readme = []
for lst in soup.find_all('li',class_ = 'repo-list-item hx_hit-repo d-flex flex-justify-start py-4 public source'):
    repo.append(DATA_PATH + "/" + lst.find("a", class_ = "v-align-middle").get('href').strip())
    repo_soup = BeautifulSoup(open(repo[i], encoding='utf8'),"html.parser")
    
    #Names
    name = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip()
    names.append(repo_soup.find('a',href = name).get_text().strip())
    
    #Languages
    if (lst.find('span', itemprop = 'programmingLanguage')):
        languages.append(lst.find('span', itemprop = 'programmingLanguage').get_text().strip())
    else:
        languages.append('none')
    
    #Stars
    stargazers = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/stargazers'
    stars.append(repo_soup.find('a',href = stargazers).get('aria-label').strip(' users starred this repository'))
    
    #Watches
    watchers = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/watchers'
    watches.append(repo_soup.find('a',href = watchers).get('aria-label').strip(' users are watching this repository'))
    
    #Forks
    forked = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/network/members'
    forks.append(repo_soup.find('a', href = forked).get('aria-label').strip(' users forked this repository'))
    
    #Commits
    commit = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/commits'
    commits.append(repo_soup.find('a', href = lambda href: href and commit in href).find('span', class_='num text-emphasized').get_text().strip())
    
    #Branches
    branch = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/branches'
    if(repo_soup.find('a', href = branch)):
        branches.append(repo_soup.find('a', href = branch).find('span', class_='num text-emphasized').get_text().strip())
    
    #Contributors
    contributor = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/graphs/contributors'
    if(repo_soup.find('a', href = branch)):
        contributors.append(repo_soup.find('a', href = contributor).find('span', class_='num text-emphasized').get_text().strip())
    
    #Issues
    issue = '/' + lst.find("a", class_ = "v-align-middle").get_text().strip() + '/issues'
    if(repo_soup.find('a', href = issue)):
        issues.append(repo_soup.find('a', href = issue).find('span', class_='Counter').get_text().strip())
    else:
        issues.append('0')
    
    #Readme
    readme.append(len(repo_soup.find('div', id='readme').get_text().strip()))
    
    i+=1

#Issues 5000"+"
for i in range(0, len(issues)): 
    if (issues[i] == '5,000+'):
        issues[i] = issues[i].replace("+","")

#Contributors "∞"
for i in range(0, len(contributors)): 
    if (contributors[i] == '∞'):
        contributors[i] = contributors[i].replace("∞","15600")

#Commits ","
for i in range(0, len(commits)): 
    commits[i] = int(commits[i].replace(",",""))

#Contributors ","
for i in range(0, len(contributors)): 
    contributors[i] = int(contributors[i].replace(",",""))

#Issues ","
for i in range(0, len(issues)): 
    issues[i] = int(issues[i].replace(",",""))

Repo_Data = pd.DataFrame({"names" : names, 
                   "languages" : languages, 
                   "stars" : stars, 
                   "watches" : watches, 
                   "forks" : forks, 
                   "commits" : commits, 
                   "branches" : branches, 
                   "contributors" : contributors, 
                   "issues" : issues, 
                   "readme" : readme})
````

````python
#DataFrame - csv
Repo_Data.to_csv('Repo_Data.csv', index=False)
````

### 2.3. Linear regression

1. We will use linear regression to predict the number of Stars based on Forks, Contributors, Issues, and Readme Length. We will try to Interpret why this is not a very good model on the basis of R-squared , F-statistic p-value, and coefficient  p-values. 
+ Then We will Develop another model which is better. Also, explaining why it is better. 

````python
#Model 1
Repo_Data_ols = sm.ols(formula="stars ~ forks + contributors + issues + readme", data=Repo_Data).fit()
Repo_Data_ols.summary()
````

````python
#Model 2
Repo_Data_ols = sm.ols(formula="stars ~ forks + watches", data=Repo_Data).fit()
Repo_Data_ols.summary()
````

**Interpretation:** For the first model, if look at the R-Squared value, we can see that the model is able to depict 22.7% of the data, and the P values of issues, readme and contributors is stating that these variables have no significance on the target variable. However, the F statistic p-value is small enough to accept the model but it is not that small either.<br>
On the other hand, if we consider the Model 2 we can say that the R-squared value is better (51%) and the P values for the coefficient (watches) is also very less, additionally, F-statistic p-value is comparitively very less.<br>
Therefore, we can say that the Mode 2 is a better model for predicting Stars.
